```python
#Question 1 - Extracting Tesla Stock Data Using yfinance
import yfinance as yf

tesla = yf.Ticker("TSLA")
tesla_history = tesla.history(period="max")
tesla_history.head()
```

```python
#Question 2 - Extracting Tesla Revenue Data Using Webscraping
import requests
from bs4 import BeautifulSoup

tesla_url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'

response = requests.get(tesla_url)
soup = BeautifulSoup(response.content, 'html.parser')

revenue_table = soup.find('table', {'class': 'historical_data_table'})
revenue_rows = revenue_table.tbody.find_all('tr')

revenue_data = []
for row in revenue_rows:
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    revenue_data.append(cols)

revenue_df = pd.DataFrame(revenue_data[1:], columns=revenue_data[0])
revenue_df.head()
```

```python
#Question 5 - Tesla Stock and Revenue Dashboard
import plotly.graph_objects as go
import yfinance as yf
import requests
from bs4 import BeautifulSoup
tesla = yf.Ticker("TSLA")
tesla_history = tesla.history(period="max")
tesla_url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'
response = requests.get(tesla_url)
soup = BeautifulSoup(response.content, 'html.parser')
revenue_table = soup.find('table', {'class': 'historical_data_table'})
revenue_rows = revenue_table.tbody.find_all('tr')
revenue_data = []
for row in revenue_rows:
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    revenue_data.append(cols)
tesla_revenue = pd.DataFrame(revenue_data[1:], columns=revenue_data[0])
fig = go.Figure()
fig.add_trace(go.Scatter(x=tesla_history.index, y=tesla_history['Close'], name='Close'))
fig.update_layout(title='Tesla Stock and Revenue Dashboard', xaxis_title='Date', yaxis_title='Price / Revenue')
fig.show()
```
